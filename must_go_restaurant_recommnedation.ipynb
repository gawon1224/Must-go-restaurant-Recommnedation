```python
# 음식점 데이터만 쓸 겁니다
df1 = df.loc[df['상권업종대분류명'] == '음식']  

# 다음과 같은 칼럼만 있으면 됩니다
df2 = df1[['상호명', '상권업종중분류명', '상권업종소분류명', '표준산업분류명', '행정동명', '위도', '경도']]

# 그 중에서도 흑석동과 상도1동만 쓸 겁니다.
df3 = df2.loc[(df['행정동명'] == '답십리1동') | (df['행정동명'] == '답십리2동')]
```

```python
# 칼럼명 단순화

df3.columns = ['name',  # 상호명
              'cate_1',  # 중분류명
              'cate_2',  # 소분류명
              'cate_3',  # 표준산업분류명
              'dong',  # 행정동명
              'lon',  # 위도
              'lat'  # 경도
              ]
```

# Step 2. 요식업 업체의 카테고리 데이터를 활용
## 코사인 유사도를 이용하여 텍스트를 기반으로 비슷함의 정도 파악하기
### 1) 카테고리 데이터들(중분류명, 소분류명, 표준산업분류명)를 하나로 합치

```python
df3['cate_mix'] = df3['cate_1'] + df3['cate_2'] + df3['cate_3']
df3['cate_mix'] = df3['cate_mix'].str.replace("/"," ")
df3
```

### 2) sklearn을 이용하여 하나로 합쳐진 카테고리 텍스트 데이터를 피쳐 백터화

```python
from sklearn.feature_extraction.text import CountVectorizer  # 피체 벡터화
from sklearn.metrics.pairwise import cosine_similarity  # 코사인 유사도
```

```python
count_vect_category = CountVectorizer(min_df=0, ngram_range=(1,2))
place_category = count_vect_category.fit_transform(df3['cate_mix']) 
place_simi_cate = cosine_similarity(place_category, place_category) 
place_simi_cate_sorted_ind = place_simi_cate.argsort()[:, ::-1]
```

```python
# 네이버 지도 검색창에 [~동 @@식당]으로 검색해 정확도를 높여야 합니다. 검색어를 미리 설정해줍시다.

df4 = df3.copy()

df4['naver_keyword'] = df4['dong'] + " " + df4['name']  # "%20"는 띄어쓰기를 의미합니다.
df4['naver_map_url'] = ''
df4['data-cid'] = ''
```

```python
# 가게 상세페이지의 URL 가져오기

for i, keyword in enumerate(df4['naver_keyword'].tolist()):
  print("이번에 검색할 키워드 :", i, f"/ {df4.shape[0] -1} 행", keyword)
  try:
    naver_map_search_url = f"https://m.map.naver.com/search2/search.naver?query={keyword}&sm=hty&style=v5"

    driver.get(naver_map_search_url)
    time.sleep(3.5)
    df4.iloc[i,-1] = driver.find_element((By.CSS_SELECTOR, "#ct > div.search_listview._content._ctList > ul > li:nth-child(1) > div.item_info > a.a_item.a_item_distance._linkSiteview").get_attribute('data-cid'))
    # 네이버 지도 시스템은 data-cid에 url 파라미터를 저장해두고 있었습니다.
    # data-cid 번호를 뽑아두었다가 기본 url 템플릿에 넣어 최종적인 url을 완성하면 됩니다.
  
  # 검색 결과가 없는 경우
  except Exception as e1:
    if "li:nth-child(1)" in str(e1):
      try:
        df4.iloc[i,-1] = driver.find_element((By.CSS_SELECTOR, "#ct > div.search_listview._content._ctList > ul > li:nth-child(1) > div.item_info > a.a_item.a_item_distance._linkSiteview").get_attribute('data-cid'))
        time.sleep(1)
      except Exception as e2:
          print(e2)
          df4.iloc[i,-1] = np.nan
          time.sleep(1)
  else:
      pass

driver.quit()
```
